```python
# No. 1 f(x) = x^3 - 2x + 1            

# Mendefinisikan f(x)
def f(x):
    # x^3 - 2x + 1     
    hasil = x**3 - 2*x + 1
    return hasil

# Mendefinisikan suatu turunan dari f(x) 
def dfdx(x):
    hasil = 3*x**2 - 2
    return hasil

#Mendefinisika suatu fungsi untuk metode Newton-Raphson
def newtonRapshon(x0, e, N):
    # Iterasi counter, dimulai dengan nilai 0
    i = 0
    
    # Variabel boolean untuk menentukan apakah akar ditemukan atau tidak, dimulai dengan nilai 1 (ditemukan) 
    # berubah apabila ada suatu kondisi suatu akar yang tidak ditemukan
    ditemukan = 1
    
    # Variabel untuk menangani division by zero, dimulai dengan nilai 0 (belum ada division by zero)
    divisionByZero = 0
    
    # Mendefinisikan xi_1 (hasil) dengan nilai x0 (apabila x0 merupakan akar persamaan)   
    xi_1 = x0
    
    while abs(f(xi_1)) > e:
        # Setiap perulangan terjadi iterasi akan bertambah
        i = i + 1
        
        # Kondisi apabila akar tidak ditemukan, jumlah perulangan melebihi maksimum iterasi yang ditentukan
        if i > N:
            ditemukan = 0
            break
        
        # Apabila turunan f(x) bernilai 0, perulangan berhenti dengan "Division by zero error"
        if dfdx(xi_1) == 0.0:
            divisionByZero = 1
            break
        
        # xi (tebakan selanjutnya) didapat dengan rumus = xi_1 (tebakan sebelumnya) - (f(xi_1)/turunan f(xi_1))
        xi = xi_1 - f(xi_1) / dfdx(xi_1)
        print("Iterasi ke-" + str(i) + ": x = " + str(xi) + ", f(x) = " + str(f(xi)) + ", Turunan f(x) = " + str(dfdx(xi)))
        
        # Hasil (xi_1) merupakan tebakan terakhir yang ditemukan dalam perulangan (xi)
        xi_1 = xi
    
    # Kondisi jika akar ditemukan, division by zero, atau akar tidak ditemukan
    if ditemukan == 1 and divisionByZero == 0:
        print('\nAkar persamaan (solusi) adalah %f' % xi_1)
    elif divisionByZero == 1:
         print('\nDivision by zero!\nAkar tidak ditemukan!')
    else:
        print('\nAkar tidak ditemukan!')
             
    
# Jalankan fungsi metode Newton-Rapshon
# Input x0 (tebakan awal), e (toleransi galat), dan N (maksimum iterasi)
x0 = input("Masukan Tebakan Pertama: ")
e = input("Toleransi galat (e): ")
N = input("Maksimum iterasi (N): ")

# Mengkonversi nilai x0 da e menjadi float
x0 = float(x0)
e = float(e)

# Mengkonversi nilai N menjadi integer
N = int(N)

# Fungsi Newton-Rapshon Dijalankan
newtonRapshon(x0, e, N)
```

```python
# No. 2 f(x) = e^x - x

# Import library math untuk fungsi eksponensial
import math

# Definisikan f(x)
def f(x):
    # e^x - x   
    hasil = math.exp(x) - x
    return hasil

# Definisikan turunan dari f(x) 
def dfdx(x):
    hasil = math.exp(x) - 1
    return hasil

# Definisikan fungsi untuk metode Newton-Raphson
def newtonRaphson(x0, e, N):
    # Iterasi counter, diawali dengan nilai 0
    i = 0
    
    # Variabel boolean untuk menentukan apakah akar ditemukan atau tidak, diawali dengan nilai 1 (ditemukan) 
    # dan akan berubah apabila ada kondisi yang membuat suatu akar tidak ditemukan
    ditemukan = 1
    
    # Variabel untuk menghandle division by zero, diawali dengan nilai 0 (belum ada division by zero)
    divisionByZero = 0
    
    # Definisikan xi_1 (hasil) dengan nilai x0 (apabila x0 merupakan akar persamaan)    
    xi_1 = x0
    
    while abs(f(xi_1)) > e:
        # Iterasi bertambah setiap perulangan terjadi
        i = i + 1
        
        # Kondisi untuk akar tidak ditemukan (jumlah perulangan (nilai i) melebihi maksimum iterasi yang ditentukan (N))
        if i > N:
            ditemukan = 0
            break
        
        # Apabila turunan f(x) bernilai 0, maka perulangan berhenti dengan "Division by zero error"
        if dfdx(xi_1) == 0.0:
            divisionByZero = 1
            break
        
        # xi (tebakan selanjutnya) didapat dengan rumus = xi_1 (tebakan sebelumnya) - (f(xi_1)/turunan f(xi_1))
        xi = xi_1 - f(xi_1) / dfdx(xi_1)
        print("Iterasi ke-" + str(i) + ": x = " + str(xi) + ", f(x) = " + str(f(xi)) + ", Turunan f(x) = " + str(dfdx(xi)))
        
        # Hasil (xi_1) merupakan tebakan terakhir yang ditemukan dalam perulangan (xi)
        xi_1 = xi
    
    # Kondisi jika akar ditemukan, division by zero, atau akar tidak ditemukan
    if ditemukan == 1 and divisionByZero == 0:
        print('\nAkar persamaan (solusi) adalah %f' % xi_1)
    elif divisionByZero == 1:
         print('\nDivision by zero!\nAkar tidak ditemukan!')
    else:
        print('\nAkar tidak ditemukan!')
             
    
# Jalankan fungsi metode Newton-Raphson
# Input x0 (tebakan awal), e (toleransi galat), dan N (maksimum iterasi)
x0 = input("Masukan Tebakan Pertama: ")
e = input("Toleransi galat (e): ")
N = input("Maksimum iterasi (N): ")

# Konversi nilai x0 da e menjadi float
x0 = float(x0)
e = float(e)

# Konversi nilai N menjadi integer
N = int(N)

# Fungsi Newton-Raphson Dijalankan
newtonRaphson(x0, e, N)
```
